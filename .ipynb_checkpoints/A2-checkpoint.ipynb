```python
import pandas as pd 
import numpy as np
# import seaborn as sns
# import warnings
# from IPython.display import display
# import os
data = pd.read_csv('Live.csv')
data
```

```python
data['status_published'] = pd.to_datetime(data['status_published'])
```

```python
data['year'] = data['status_published'].dt.year
data['month'] = data['status_published'].dt.month
data['day'] = data['status_published'].dt.day
print(data.dtypes)
```

```python
# check missing value
data.isnull().sum()
```

```python
data.drop(labels=['Column1', 'Column2', 'Column3','Column4'], axis=1, inplace=True)
data
```

```python
# sanity check
print(data.describe().transpose()) 
```

```python
def sanity_check_name(a): 
    print("Sanity check for " + a + " column")
def sanity_check(a,b):
    for h in a:
        if b <= h : # a condition
            print True     
        else:
            print ("Something not True here and it is: ")
            print h
sanity_check_name("num_reactions")
sanity_check(data["num_reactions"],0)
sanity_check_name("num_comments")
sanity_check(data["num_comments"],0)
sanity_check_name("num_shares")
sanity_check(data["num_shares"],0)
sanity_check_name("num_likes")
sanity_check(data["num_likes"],0)
sanity_check_name("num_loves")
sanity_check(data["num_loves"],0)
sanity_check_name("num_wows")
sanity_check(data["num_wows"],0)
sanity_check_name("num_hahas")
sanity_check(data["num_hahas"],0)
sanity_check_name("num_sads")
sanity_check(data["num_sads"],0)
sanity_check_name("num_angrys")
sanity_check(data["num_angrys"],0)
```

```python
data['status_type'].value_counts()
```

```python
mask_count_photo = data['status_type'] == 'photo'
mask_count_video = data['status_type'] == 'video'
mask_count_status = data['status_type'] == 'status'
mask_count_link = data['status_type'] == 'link'

photo_count_nr = data.dropna().loc[mask_count_photo, 'num_reactions'].sum()
video_count_nr = data.dropna().loc[mask_count_video, 'num_reactions'].sum()
status_count_nr = data.dropna().loc[mask_count_status, 'num_reactions'].sum()
link_count_nr = data.dropna().loc[mask_count_link, 'num_reactions'].sum()

photo_count_sh = data.dropna().loc[mask_count_photo, 'num_shares'].sum()
video_count_sh = data.dropna().loc[mask_count_video, 'num_shares'].sum()
status_count_sh = data.dropna().loc[mask_count_status, 'num_shares'].sum()
link_count_sh = data.dropna().loc[mask_count_link, 'num_shares'].sum()

photo_count_like = data.dropna().loc[mask_count_photo, 'num_likes'].sum()
video_count_like = data.dropna().loc[mask_count_video, 'num_likes'].sum()
status_count_like = data.dropna().loc[mask_count_status, 'num_likes'].sum()
link_count_like = data.dropna().loc[mask_count_link, 'num_likes'].sum()

photo_count_cmt = data.dropna().loc[mask_count_photo, 'num_comments'].sum()
video_count_cmt = data.dropna().loc[mask_count_video, 'num_comments'].sum()
status_count_cmt = data.dropna().loc[mask_count_status, 'num_comments'].sum()
link_count_cmt = data.dropna().loc[mask_count_link, 'num_comments'].sum()

count_nr = [photo_count_nr, video_count_nr, status_count_nr, link_count_nr]
count_sh = [photo_count_sh, video_count_sh, status_count_sh, link_count_sh]
count_like = [photo_count_like, video_count_like, status_count_like, link_count_like]
count_cmt = [photo_count_cmt, video_count_cmt, status_count_cmt, link_count_cmt]
```

```python
plt.figure (figsize=(12,8))
plt.bar(range(4), count_nr)
plt.xticks(range(4),['photo','video','status','link'])
plt.title('Number of reaction depend on status types', color = 'navy')
plt.xlabel('Status types')
plt.ylabel('Amount of reaction')
```

```python
plt.figure (figsize=(12,8))
plt.bar(range(4), count_sh)
plt.xticks(range(4),['photo','video','status','link'])
plt.title('Time of share depend on status types', color = 'navy')
plt.xlabel('Status types')
plt.ylabel('Time of share')
```

```python
plt.figure (figsize=(12,8))
plt.bar(range(4), count_like)
plt.xticks(range(4),['photo','video','status','link'])
plt.title('Number of likes depend on status types', color = 'navy')
plt.xlabel('Status types')
plt.ylabel('Amount of like')
```

```python
plt.figure (figsize=(12,8))
plt.bar(range(4), count_cmt)
plt.xticks(range(4),['photo','video','status','link'])
plt.title('Number of comments depend on status types', color = 'navy')
plt.xlabel('Status types')
plt.ylabel('Amount of comment')
```
